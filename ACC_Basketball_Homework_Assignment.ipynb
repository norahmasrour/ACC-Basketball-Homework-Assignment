```python
school_points = df.groupby('School')['PTS'].sum()
school_points
```

```python
top_3_schools = school_points.sort_values(ascending=False).head(3)
top_3_schools
```

```python
top_5_players = df[['Player', 'PTS']].sort_values(by='PTS', ascending=False).head(5)
top_5_players
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.barh(top_5_players['Player'], top_5_players['PTS'], color='brown')
plt.xlabel('Total Points')
plt.ylabel('Player')
plt.title('Top 5 Players by Total Points Scored')
plt.gca().invert_yaxis()  # Invert the y-axis to have the highest value on top
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.scatter(df['FG%'], df['WS'], color='blue')
plt.xlabel('Field Goal Percentage (FG%)')
plt.ylabel('Win Shares (WS)')
plt.title('Correlation between FG% and WS')
```
